```python
# 1) Now loop through all the DICOM files and store them in a 3D numpy array.
# Hint: You can either store them in a list first or read the dimensions of a
# single image slice to properly create the 3D numpy array.
# Hint 2: os.listdir(DIR) gives a list of filenames in a directory.
# Hint 2b: This list is not sorted - make sure you sort it.
# Hint 3: The dcmread function loads a single DICOM file.
# Hint 4: You can then use .pixel_array to access the image data.
```

```python
import os
import numpy as np
import pydicom
```

```python
from pydicom import dcmread
```

```python
# TODO: YOUR CODE FOR LOADING THE VOLUME AS A 3D NUMPY ARRAY
```

```python
dir_path = "/content/ct/"
```

```python
os.listdir(dir_path)
```

```python
file_list = sort(os.listdir(dir_path))
```

```python
first_dir_path = dcmread(os.path.join(dir_path, file_list[0]))
shape = first_dir_path.pixel_array.shape
```

```python
data = np.empty((shape[0], shape[1], len(file_list)), dtype=first_dir_path.pixel_array.dtype)
```

```python
for i, filename in enumerate(file_list):
    dcm = dcmread(os.path.join(dir_path, filename))
    data[:, :, i] = dcm.pixel_array
```

**Bonus**: Create segmentations (label maps) for the volume using thresholding HU! [33 Points]

```python
# Similar to Window/Level adjustment for visualization, we can threshold
# the volume to highlight the following components using the Hounsfield Units:
# 1) Fat
# 2) Soft Tissue
# 3) Bones
#
# Please create 3 segmentation masks for these structures.
# Then, please visualize each 3 slices per structure to showcase the segmentation.
# Hint: As a reminder, the following code allows thresholding of a numpy array.
#   new_mask = imagevolume.copy()
#   new_mask[new_mask < XXX] = 0
# Hint2: You might need to cast new_mask to int16 not uint16.
```

```python
# TODO: YOUR CODE TO SEGMENT FAT
```

```python
# TODO: YOUR CODE TO SEGMENT FAT
# Threshold values from https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3163666/
fat_min = 500
fat_max = 1000
# Create a copy of the volume to modify
fat_mask = data.copy()
# Threshold the copy to create the segmentation mask
fat_mask[(fat_mask < fat_min) | (fat_mask > fat_max)] = 0
fat_mask[fat_mask != 0] = 1
# Visualize three slices
plt.figure()
for i in range(100, 101):
    plt.subplot(1, 1, i - 99)
    plt.imshow(fat_mask[i], cmap='gray')
plt.show()
```

```python
# TODO: YOUR CODE TO SEGMENT SOFT TISSUE
```

```python
# TODO: YOUR CODE TO SEGMENT SOFT TISSUE
# Threshold values from https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3163666/
soft_tissue_min = 1000
soft_tissue_max = 1200
# Create a copy of the volume to modify
soft_tissue_mask = data.copy()
# Threshold the copy to create the segmentation mask
soft_tissue_mask[(soft_tissue_mask < soft_tissue_min) | (soft_tissue_mask > soft_tissue_max)] = 0
soft_tissue_mask[soft_tissue_mask != 0] = 1
# Visualize three slices
plt.figure()
for i in range(100, 101):
    plt.subplot(1, 1, i - 99)
    plt.imshow(soft_tissue_mask[i], cmap='gray')
plt.show()
```

```python
# TODO: YOUR CODE TO SEGMENT BONES
```

```python
# TODO: YOUR CODE TO SEGMENT BONES
# Threshold values from https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3163666/
bone_min = 200
bone_max = 3000
# Create a copy of the volume to modify
bone_mask = data.copy()
# Threshold the copy to create the segmentation mask
bone_mask[(bone_mask < bone_min) | (bone_mask > bone_max)] = 0
bone_mask[bone_mask != 0] = 1
# Visualize three slices
plt.figure()
for i in range(100, 101):
    plt.subplot(1, 1, i - 99)
    plt.imshow(bone_mask[i], cmap='gray')
plt.show()
```
